```python
from tkinter import *
from selenium import webdriver #criar um navegador
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By #Localizar elementos (os itens de um site)
from selenium.webdriver.common.keys import Keys # Permite clicar em teclas do teclado
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd


def cotar():
    fundos = []
    digitar_fundo = input("Digite um ou mais códigos dos fundos imobiliários ou ações. Para acessar a pesquisa, tecle 'ENTER' com a caixa vazia: ").upper()
    while digitar_fundo != '':
        fundos.append(digitar_fundo)
        digitar_fundo = input("Próximo FUNDO ou AÇÃO. Para acessar a pesquisa, tecle 'ENTER' com a caixa vazia: ").upper()
    return fundos

def simular():
    return


window = Tk()

window.geometry("996x613")
window.configure(bg = "#FFFFFF")
canvas = Canvas(
    window,
    bg = "#FFFFFF",
    height = 613,
    width = 996,
    bd = 0,
    highlightthickness = 0,
    relief = "ridge")
canvas.place(x = 0, y = 0)

background_img = PhotoImage(file = f"bg.png")
background = canvas.create_image(
    498.0, 306.5,
    image=background_img)

img0 = PhotoImage(file = f"btsim.png")
b0 = Button(
    image = img0,
    borderwidth = 0,
    highlightthickness = 0,
    command = simular,
    relief = "flat")

b0.place(
    x = 397, y = 218,
    width = 208,
    height = 79)

img1 = PhotoImage(file = f"bt.png")
b1 = Button(
    image = img1,
    borderwidth = 0,
    highlightthickness = 0,
    command = cotar,
    relief = "flat")

b1.place(
    x = 149, y = 218,
    width = 208,
    height = 79)


window.resizable(False, False)
window.mainloop()
```

```python
servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
navegador.get('https://www.fundsexplorer.com.br/ranking')
tabela_fundos = navegador.find_element(By.ID, 'table-ranking').text
cotacao_fii = []
cotacao_acoes = []
for fundo in fundos:
    if fundo in tabela_fundos:
        navegador.get('https://www.fundsexplorer.com.br/funds'+ '/' + fundo) 
        preco = navegador.find_element(By.CLASS_NAME, 'price').text
        ultimo_rendimento = navegador.find_element(By.XPATH, '//*[@id="main-indicators-carousel"]/div/div/div[2]/span[2]').text
        dy = navegador.find_element(By.XPATH, '//*[@id="main-indicators-carousel"]/div/div/div[3]/span[2]').text
        dy_3m = navegador.find_element(By.XPATH,'//*[@id="dividends"]/div/div/div[2]/div[1]/div/div[2]/table/tbody/tr[2]/td[3]').text
        dy_6m = navegador.find_element(By.XPATH,'//*[@id="dividends"]/div/div/div[2]/div[1]/div/div[2]/table/tbody/tr[2]/td[4]').text
        dy_12m =navegador.find_element(By.XPATH,'//*[@id="dividends"]/div/div/div[2]/div[1]/div/div[2]/table/tbody/tr[2]/td[5]').text
        nome = navegador.find_element(By.XPATH, '//*[@id="head"]/div/div/div/h3').text
        p_vp = navegador.find_element(By.XPATH, '//*[@id="main-indicators-carousel"]/div/div/div[7]/span[2]').text
        rent_mes = navegador.find_element(By.XPATH, '//*[@id="main-indicators-carousel"]/div/div/div[6]/span[2]').text
        liq_diaria = navegador.find_element(By.XPATH, '//*[@id="main-indicators-carousel"]/div/div/div[1]/span[2]').text
        patr_liquido = navegador.find_element(By.XPATH, '//*[@id="main-indicators-carousel"]/div/div/div[4]/span[2]').text
        segmento = navegador.find_element(By.XPATH, '//*[@id="basic-infos"]/div/div/div[2]/div/div[2]/ul/li[4]/div[2]/span[2]').text
        cotacao_fii.append([fundo, preco.replace('R$ ', ''), ultimo_rendimento.replace('R$ ', ''), dy.replace('%', ''), dy_3m.replace('%', ''), dy_6m.replace('%', ''), dy_12m.replace('%', ''), p_vp, rent_mes.replace('%', ''), liq_diaria, patr_liquido.replace('R$ ', ''), segmento, nome])
    else:
        navegador.get('https://statusinvest.com.br/acoes/'+ fundo)
        cot = navegador.find_element(By.XPATH, '//*[@id="main-2"]/div[2]/div/div[1]/div/div[1]/div/div[1]/strong').text
        variacao_dia = navegador.find_element(By.XPATH, '//*[@id="main-2"]/div[2]/div/div[1]/div/div[1]/div/div[2]/span/b').text
        dy_acoes = navegador.find_element(By.XPATH, '//*[@id="main-2"]/div[2]/div/div[1]/div/div[4]/div/div[1]/strong').text
        nome_empresa = navegador.find_element(By.XPATH, '//*[@id="main-header"]/div[2]/div/div[1]/h1/small').text
        valorizacao_12m = navegador.find_element(By.XPATH, '//*[@id="main-2"]/div[2]/div/div[1]/div/div[5]/div/div[1]/strong').text
        valorizacao_mes = navegador.find_element(By.XPATH, '//*[@id="main-2"]/div[2]/div/div[1]/div/div[5]/div/div[2]/div/span[2]/b').text
        min_52_semanas = navegador.find_element(By.XPATH,'//*[@id="main-2"]/div[2]/div/div[1]/div/div[2]/div/div[1]/strong').text
        max_52_semanas = navegador.find_element(By.XPATH,'//*[@id="main-2"]/div[2]/div/div[1]/div/div[3]/div/div[1]/strong').text
        variacao_dia_anterior = navegador.find_element(By.XPATH,'//*[@id="main-2"]/div[2]/div/div[1]/div/div[1]/div/div[2]/span/b').text
        part_IBOV = navegador.find_element(By.XPATH, '//*[@id="main-2"]/div[2]/div/div[5]/div/div/div[4]/div/a/div/div/strong').text
        patr_liquido = navegador.find_element(By.XPATH,'//*[@id="company-section"]/div[1]/div/div[2]/div[1]/div/div/strong').text
        ativos = navegador.find_element(By.XPATH,'//*[@id="company-section"]/div[1]/div/div[2]/div[2]/div/div/strong').text
        divida_liquida = navegador.find_element(By.XPATH,'//*[@id="company-section"]/div[1]/div/div[2]/div[6]/div/div/strong').text
        valor_de_mercado = navegador.find_element(By.XPATH,'//*[@id="company-section"]/div[1]/div/div[2]/div[7]/div/div/strong').text
        cotacao_acoes.append([fundo, cot, variacao_dia.replace('%', ''), dy_acoes, nome_empresa, variacao_dia_anterior.replace('%', ''), valorizacao_mes.replace('%', ''), valorizacao_12m.replace('%', ''), min_52_semanas, max_52_semanas, part_IBOV, patr_liquido, ativos, divida_liquida, valor_de_mercado])

cotacao_fii = tuple(cotacao_fii)
cotacao_fundos_acoes = tuple(cotacao_acoes)

cotacao_fundos_fii = pd.DataFrame(cotacao_fii)
cotacao_fundos_fii = cotacao_fundos_fii.rename(columns={0: 'Fundo', 1: 'Cotação Atual (R$)', 2: 'Último Rendimento (R$)', 3: 'Dividend Yield (%)', 4: 'Dividend Yield (3M) (%)', 5: 'Dividend Yield (6M) (%)', 6: 'Dividend Yield (12M) (%)',7: 'P/VP', 8: 'Rentabilidade no Mês (%)', 9: 'Liquidez Diária', 10: 'Patrimônio Líquido (R$)' , 11: 'Segmento', 12:'Nome do Fundo'})
cotacao_fundos_fii['Cotação Atual (R$)'] = cotacao_fundos_fii['Cotação Atual (R$)'].str.replace(',', '.')
cotacao_fundos_fii['Último Rendimento (R$)'] = cotacao_fundos_fii['Último Rendimento (R$)'].str.replace(',', '.')
cotacao_fundos_fii['Dividend Yield (%)'] = cotacao_fundos_fii['Dividend Yield (%)'].str.replace(',', '.')
cotacao_fundos_fii['Dividend Yield (3M) (%)'] = cotacao_fundos_fii['Dividend Yield (3M) (%)'].str.replace(',', '.')
cotacao_fundos_fii['Dividend Yield (6M) (%)'] = cotacao_fundos_fii['Dividend Yield (6M) (%)'].str.replace(',', '.')
cotacao_fundos_fii['Dividend Yield (12M) (%)'] = cotacao_fundos_fii['Dividend Yield (12M) (%)'].str.replace(',', '.')
cotacao_fundos_fii['P/VP'] = cotacao_fundos_fii['P/VP'].str.replace(',', '.')
cotacao_fundos_fii['Rentabilidade no Mês (%)'] = cotacao_fundos_fii['Rentabilidade no Mês (%)'].str.replace(',', '.')
cotacao_fundos_fii['Patrimônio Líquido (R$)'] = cotacao_fundos_fii['Patrimônio Líquido (R$)'].str.replace(',', '.')
cotacao_fundos_fii

cotacao_fundos_fii['Cotação Atual (R$)'] = cotacao_fundos_fii['Cotação Atual (R$)'].astype(float)
cotacao_fundos_fii['Último Rendimento (R$)'] = cotacao_fundos_fii['Último Rendimento (R$)'].astype(float)


fig = px.bar(cotacao_fundos_fii, x='Fundo', y=cotacao_fundos_fii['Último Rendimento (R$)'], 
             text='Último Rendimento (R$)', title="Rendimentos Por Fundo")
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.show()

fig = px.bar(cotacao_fundos_fii, x='Fundo', y='Cotação Atual (R$)', 
             text='Cotação Atual (R$)', title="Cotações do Dia", color='Último Rendimento (R$)')
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.show()


cotacao_fundos_acoes = pd.DataFrame(cotacao_fundos_acoes)
cotacao_fundos_acoes = cotacao_fundos_acoes.rename(columns={0: 'Ticker da Ação', 1: 'Cotação Atual (R$)', 2: 'Variação do Dia (%)', 3: 'Dividend Yield (%)', 4: 'Nome da Empresa (%)', 5:'Var. Dia Anterior (%)', 6: 'Valorização no Mês (%)', 7: 'Valorização (12M) (%)', 8: 'Min(52sem)(R$)', 9: 'Max(52sem)(R$)', 10: 'Part.IBOV (%)', 11: 'Patrimônio Líquido (R$)', 12: 'Ativos (R$)', 13: 'Dívida Líquida (R$)', 14: 'Valor de Mercado (R$)' })

cotacao_fundos_acoes['Cotação Atual (R$)'] = cotacao_fundos_acoes['Cotação Atual (R$)'].str.replace(',', '.')
cotacao_fundos_acoes['Variação do Dia (%)'] = cotacao_fundos_acoes['Variação do Dia (%)'].str.replace(',', '.')
cotacao_fundos_acoes['Dividend Yield (%)'] = cotacao_fundos_acoes['Dividend Yield (%)'].str.replace(',', '.')
cotacao_fundos_acoes['Var. Dia Anterior (%)'] = cotacao_fundos_acoes['Var. Dia Anterior (%)'].str.replace(',', '.')
cotacao_fundos_acoes['Valorização no Mês (%)'] = cotacao_fundos_acoes['Valorização no Mês (%)'].str.replace(',', '.')
cotacao_fundos_acoes['Valorização (12M) (%)'] = cotacao_fundos_acoes['Valorização (12M) (%)'].str.replace(',', '.')
cotacao_fundos_acoes['Valorização (12M) (%)'] = cotacao_fundos_acoes['Valorização (12M) (%)'].str.replace(',', '.')
cotacao_fundos_acoes['Min(52sem)(R$)'] = cotacao_fundos_acoes['Min(52sem)(R$)'].str.replace(',', '.')
cotacao_fundos_acoes['Part.IBOV (%)'] = cotacao_fundos_acoes['Part.IBOV (%)'].str.replace(',', '.')
cotacao_fundos_acoes['Max(52sem)(R$)'] = cotacao_fundos_acoes['Max(52sem)(R$)'].str.replace(',', '.')


cotacao_fundos_acoes

cotacao_fundos_acoes['Cotação Atual (R$)'] = cotacao_fundos_acoes['Cotação Atual (R$)'].astype(float)
cotacao_fundos_acoes['Min(52sem)(R$)'] = cotacao_fundos_acoes['Min(52sem)(R$)'].astype(float)
cotacao_fundos_acoes['Max(52sem)(R$)'] = cotacao_fundos_acoes['Max(52sem)(R$)'].astype(float)

fig = px.bar(cotacao_fundos_acoes, x='Ticker da Ação', y='Cotação Atual (R$)', 
             text='Cotação Atual (R$)', title="Cotações do Dia", color='Cotação Atual (R$)')
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.show()

fig = go.Figure(data=[go.Bar(name='ATUAL', x=cotacao_fundos_acoes['Ticker da Ação'], y=cotacao_fundos_acoes['Cotação Atual (R$)']), 
                      go.Bar(name='MIN', x=cotacao_fundos_acoes['Ticker da Ação'], y=cotacao_fundos_acoes['Min(52sem)(R$)']),
                      go.Bar(name='MAX', x=cotacao_fundos_acoes['Ticker da Ação'], y=cotacao_fundos_acoes['Max(52sem)(R$)'])])

fig.update_traces(textfont_size=8, textangle=0, textposition="outside", cliponaxis=False)
              
fig.show()


         
```
